```python

# library imports
import pandas as pd

# local file imports
import src.util.dataset_loader as dsl
```

```python
a = dsl.load_datasets("data")

get_num_cols = lambda x: len(x.keys())
get_num_rows = lambda x: x[list(x.keys())[0]].shape[0]

rows = []
for ds, values in a:
    rows.append([ds.name, get_num_cols(values), get_num_rows(values)])

dataset_descriptor = pd.DataFrame(rows, columns=["name", "num_columns", "num_rows"])
    
```

```python
# dataset_descriptor.sort_values("num_columns", ascending=False, inplace=True)
dataset_descriptor.sort_index(inplace=True)

display(dataset_descriptor)
```
